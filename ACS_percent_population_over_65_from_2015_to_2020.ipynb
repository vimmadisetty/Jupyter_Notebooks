```python
import pandas as pd
import csv
```

```python
#reading the CSV file
df1=pd.read_excel('ACS_All_population_2015_to_2020_Subjected_data.xlsx')
df1.head()
```

```python
#reading the CSV file
df2=pd.read_excel('ACS_2015_to_2020_Population_Over_65.xlsx')
df2.head()
```

```python
#renmaing the column names
df2.rename(columns={'DP05_0025E (Estimate!!SEX AND AGE!!65 years and over)':'Total_population_over_65','DP05_0026E (Estimate!!SEX AND AGE!!65 years and over!!Male)':'Male_population_over_65','DP05_0027E (Estimate!!SEX AND AGE!!65 years and over!!Female)':'Female_population_over_65'},inplace=True)
df2.head()
```

```python
#drop a column
df1=df1.drop('S0101_C02_001E (Total_Estimated_Male_Population)',axis=1)
df1.head()
```

```python
#drop a column
df1=df1.drop('S0101_C03_001E (Total_Estimated_Female_Population)',axis=1)
#renmaing the column name
df1.rename(columns={'S0101_C01_001E (Total_Estimated_Population)':'Total_population'},inplace=True)
df1.head()
```

```python
#drop a column
df2 = df2.drop('Male_population_over_65',axis=1)
#drop a column
df2 = df2.drop('Female_population_over_65',axis=1)
df2.head()
```

```python
#merging two dataframes based on county,state and year
df3=pd.merge(df1,df2,on=['County','State','Year'])
df3.head()
```

```python
#gives the details about datatypes and non-null count of the dataframe
df3.info()
```

```python
#mathematical calculation 
df3['Percent_population_over_65']=(df3['Total_population_over_65']*100)/df3['Total_population']
df3.head()
```

```python
#pivoting a table and reseting the index
df3=df3.pivot_table(index=['County','State'],values=['Total_population','Total_population_over_65','Percent_population_over_65'],columns='Year').reset_index()
df3.head()
```

```python
#converting dataframe into csv
df3.to_csv('ACS_percent_population_over_65_from_2015_to_2020.csv',index=True)
```
