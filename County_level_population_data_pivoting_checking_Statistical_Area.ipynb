```python
import pandas as pd
import csv
```

```python
#Reading the data 
df1=pd.read_excel('ACS_county_level_population_data.xlsx')
df1.head()
```

```python
#drop the unnecessary columns
df1=df1.drop('S0101_C02_001E (Total_Estimated_Male_Population)',axis=1)
df1=df1.drop('S0101_C03_001E (Total_Estimated_Female_Population)',axis=1)
df1.head()
```

```python
#pivoting the table based on the year
df1=df1.pivot_table(index=['County','State'],values=['S0101_C01_001E (Total_Estimated_Population)'],columns='Year').reset_index()
df1.head()
```

```python
#df1['State'].str.strip()
df1.head()
```

```python
#read the data from excel
df2=pd.read_excel('Statistical_areas.xlsx')
df2.head()
```

```python
#concatinating the state and county to create a new key
df1['Key']=df1['State']+'-'+df1['County']
df1.head()
```

```python
#df1['Key'].str.strip()
df1.head()
```

```python
#concatinating the state and county to create a new key
df2['Key']=df2['State']+'-'+df2['County']
df2.head()
```

```python
#sorting the values 
df1=df1.sort_values(by=['State'])
df1.head()
```

```python
df2.to_csv('df2_keys.csv',index=False)
df1.to_csv('df1_keys.csv',index=False)
```

```python
df3=pd.read_csv('df1_keys.csv')
df3.head()
```

```python
df4=pd.read_csv('df2_keys.csv')
df4.head()
```

```python
#merge the two dataframes based on the key column in both the dataframes
merge=pd.merge(df3,df4,left_on='Key',right_on='Key',how='inner')
merge.head()
```

```python
merge.to_csv('Final_inner_merge_data_11-08-22.csv',index=False)
```

```python
#getting the unmatched data from dataframe 3 
unmatched_data=df3[(~df3.Key.isin(df4.Key))]
unmatched_data.head()
```

```python
unmatched_data.to_csv("unmatched_data_check.csv",index=False)
```
