```python
import pandas as pd
import csv
```

```python
#import the data from excel to a dataframe
df1=pd.read_excel('ACS_All_population_2015_to_2020_Subjected_data.xlsx')
```

```python
#This will print the first few line of the dataframe
df1.head()
```

```python
#performing the arithmetic operations on the dataset and storing them in a new column
df1['Percent_Male_population']= (df1['S0101_C02_001E (Total_Estimated_Male_Population)']*100)/df1['S0101_C01_001E (Total_Estimated_Population)']
df1.head()
```

```python
#performing the arithmetic operations on the dataset and storing them in a new column
df1['Percent_Female_population']= (df1['S0101_C03_001E (Total_Estimated_Female_Population)']*100)/df1['S0101_C01_001E (Total_Estimated_Population)']
df1.head()
```

```python
df1.to_csv('ACS_updated_with_percent_male_female.csv',index=False)
```

```python
df1.head()
```

```python
#dropping the unncessary columns, axis=1 represent a column
df1=df1.drop('S0101_C02_001E (Total_Estimated_Male_Population)',axis=1)
df1.head()
```

```python
#dropping the unncessary columns, axis=1 represent a column
df1=df1.drop('S0101_C03_001E (Total_Estimated_Female_Population)',axis=1)
df1=df1.drop('Id_number',axis=1)
df1.head()
```

```python
#converting the rows data into column by using the pivot_table function
df1=df1.pivot_table(index=['County','State'],values=['S0101_C01_001E (Total_Estimated_Population)','Percent_Male_population','Percent_Female_population'],columns='Year').reset_index()
df1.head()
```

```python
df1.to_csv('updated_ACS_population.csv',index=False)
```
