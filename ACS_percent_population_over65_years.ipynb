```python
import pandas as pd
import csv
```

```python
#import the data from excel to a dataframe
df1=pd.read_excel('ACS_population_over_65_from_2015_to_2020.xlsx')
#This will print the first few line of the dataframe
df1.head()
```

```python
#dropping the unncessary columns, axis=1 represent a column
df1=df1.drop('Total_Male_over_65_years',axis=1)
#dropping the unncessary columns, axis=1 represent a column
df1=df1.drop('Total_Female_population_over_65_years',axis=1)
df1.head()
```

```python
#performing the arithmetic operations on the dataset and storing them in a new column
df1['Percent_population_over_65_years']=(df1['Total_Population_over_65_years']*100)/df1['Total_population']
df1.head()
```

```python
#converting the rows data into column by using the pivot_table function
df1=df1.pivot_table(index=['County','State'],values=['Percent_population_over_65_years','Total_population','Total_Population_over_65_years'],columns='Year').reset_index()
df1.head()
```

```python
#storing the data present in the dataframe in a excel while
df1.to_excel('final_updated_ACS_percent_population_over_65years.xlsx')
```
