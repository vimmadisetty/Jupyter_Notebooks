```python
import pandas as pd
import csv
```

```python
df1=pd.read_csv('sample_csv.csv',encoding='cp1252')
df1.head()
```

```python
df1.info()
```

```python
#appending zeros to the leading numbers based on the requirement using the zfill()
df1['state_no']=df1['state_no'].apply(lambda x: x.zfill(2))
df1['county_no']=df1['county_no'].apply(lambda x: x.zfill(3))
df1['tract']=df1['tract'].apply(lambda x: x.zfill(6))
df1.head()
```

```python
df1.info()
```

```python
#concatinating the state,county,tract codes to get the full length census tract code which is 11 digits
df1['Full_census_tract_code']=df1['state_no']+df1['county_no']+df1['tract']
df1.head()
```

```python
#formating the data to store the appended zeros at the leading nu bers so that we can see the numbers as expected when exporting the data file
df1['state_no'] = df1['state_no'].apply('="{}"'.format)
df1['county_no'] = df1['county_no'].apply('="{}"'.format)
df1['tract'] = df1['tract'].apply('="{}"'.format)
df1['Full_census_tract_code'] = df1['Full_census_tract_code'].apply('="{}"'.format)
df1.head()
```

```python
df1.to_excel('Census_tracts_data_2020_full_Census_tract_code.xlsx',index=False)
```
