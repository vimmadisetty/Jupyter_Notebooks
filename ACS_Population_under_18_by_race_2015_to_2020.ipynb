```python
import pandas as pd
import csv
```

```python
#import the data from excel to a dataframe
df1=pd.read_excel('ACS_All_population_2015_to_2020_Subjected_data.xlsx')
#This will print the first few line of the dataframe
df1.head()
```

```python
#import the data from excel to a dataframe
df2=pd.read_excel('ACS_under18_Poplation_from_2015_to_2020.xlsx')
#This will print the first few line of the dataframe
df2.head()
```

```python
# This will add all the column names of the dataframe into a list and displays them 
df2.columns.to_list
```

```python
#dropping the unncessary columns, axis=1 represent a column
df1=df1.drop('S0101_C02_001E (Total_Estimated_Male_Population)',axis=1)
#dropping the unncessary columns, axis=1 represent a column
df1=df1.drop('S0101_C03_001E (Total_Estimated_Female_Population)',axis=1)
df1.head()
```

```python
#merging 2 dataframes based on County, State and Year which are common columns in both the data framesand storing them in df3
df3=pd.merge(df1,df2,on=['County','State','Year'])
df3.head()
```

```python
# This will add all the column names of the dataframe into a list and displays them 
df3.columns.to_list
```

```python
#performing the arithmetic operations on the dataset and storing them in a new column
df3['Total_white_population_under_18']=df3['Total_Male_Under_18years (WHITE ALONE)']+df3['Total_Female_Under_18years (WHITE ALONE)']
df3['Total_BLACK_OR_AFRICAN_AMERICAN_ALONE_population_under_18']=df3['Total_Male_Under_18years (BLACK OR AFRICAN AMERICAN ALONE)']+df3['Total_Female_Under_18years (BLACK OR AFRICAN AMERICAN ALONE)']
df3['Total_AMERICAN_INDIAN_AND_ALASKA_NATIVE_ALONE_population_under_18']=df3['Total_Male_under_18Year(AMERICAN INDIAN AND ALASKA NATIVE ALONE)']+df3['Total_Female_under_18Year(AMERICAN INDIAN AND ALASKA NATIVE ALONE)']
df3['Total_ASIAN_ALONE_population_under_18']=df3['Total_Male_Under_18years(ASIAN ALONE)']+df3['Total_Female_Under_18years(ASIAN ALONE)']
df3['Total_NATIVE_HAWAIIAN_AND_OTHER_PACIFIC_ISLANDER_ALONE_population_under_18']=df3['Total_Male_Under_18years(NATIVE HAWAIIAN AND OTHER PACIFIC ISLANDER ALONE)']+df3['Total_Female_Under_18years(NATIVE HAWAIIAN AND OTHER PACIFIC ISLANDER ALONE)']
df3['Total_SOME_OTHER_RACE_ALONE_population_under_18']=df3['Total_Male_Under_18years(SOME OTHER RACE ALONE)']+df3['Total_Female_Under_18years(SOME OTHER RACE ALONE)']
df3['Total_TWO_OR_MORE_RACES_population_under_18']=df3['Total_Male_Under_18years(TWO OR MORE RACES)']+df3['Total_Female_Under_18years(TWO OR MORE RACES)']
df3['Total_WHITE_ALONE_NOT_HISPANIC_OR_LATINO_population_under_18']=df3['Total_Male_Under_18years(WHITE ALONE, NOT HISPANIC OR LATINO)']+df3['Total_Female_Under_18years(WHITE ALONE, NOT HISPANIC OR LATINO)']
df3['Total_HISPANIC_OR_LATINO_population_under_18']=df3['Total_Male_under_18years(HISPANIC OR LATINO)']+df3['Total_Female_under_18years(HISPANIC OR LATINO)']
df3.head()
```

```python
#performing the arithmetic operations on the dataset and storing them in a new column
df3['percent_white_population_under_18']=(df3['Total_white_population_under_18']*100)/df3['S0101_C01_001E (Total_Estimated_Population)']
df3['Percent_BLACK_OR_AFRICAN_AMERICAN_ALONE_population_under_18']=(df3['Total_BLACK_OR_AFRICAN_AMERICAN_ALONE_population_under_18']*100)/df3['S0101_C01_001E (Total_Estimated_Population)']
df3['percent_AMERICAN_INDIAN_AND_ALASKA_NATIVE_ALONE_population_under_18']=(df3['Total_AMERICAN_INDIAN_AND_ALASKA_NATIVE_ALONE_population_under_18']*100)/df3['S0101_C01_001E (Total_Estimated_Population)']
df3['percent_ASIAN_ALONE_population_under_18']=(df3['Total_ASIAN_ALONE_population_under_18']*100)/df3['S0101_C01_001E (Total_Estimated_Population)']
df3['percent_NATIVE_HAWAIIAN_AND_OTHER_PACIFIC_ISLANDER_ALONE_population_under_18']=(df3['Total_NATIVE_HAWAIIAN_AND_OTHER_PACIFIC_ISLANDER_ALONE_population_under_18']*100)/df3['S0101_C01_001E (Total_Estimated_Population)']
df3['percent_SOME_OTHER_RACE_ALONE_population_under_18']=(df3['Total_SOME_OTHER_RACE_ALONE_population_under_18']*100)/df3['S0101_C01_001E (Total_Estimated_Population)']
df3['percent_TWO_OR_MORE_RACES_population_under_18']=(df3['Total_TWO_OR_MORE_RACES_population_under_18']*100)/df3['S0101_C01_001E (Total_Estimated_Population)']
df3['percent_WHITE_ALONE_NOT_HISPANIC_OR_LATINO_population_under_18']=(df3['Total_WHITE_ALONE_NOT_HISPANIC_OR_LATINO_population_under_18']*100)/df3['S0101_C01_001E (Total_Estimated_Population)']
df3['percent_HISPANIC_OR_LATINO_population_under_18']=(df3['Total_HISPANIC_OR_LATINO_population_under_18']*100)/df3['S0101_C01_001E (Total_Estimated_Population)']
```

```python
#storing the data present in the dataframe in a csv while
df3.to_csv('updated_under_18_race_population_percentage.csv',index=False)
```

```python
#import the data from csv to a dataframe
df4=pd.read_csv('updated_under_18_race_population_percentage.csv')
df4.head()
```

```python
df4.columns.to_list
```

```python
#converting the rows data into column by using the pivot_table function
df4=df4.pivot_table(index=['County','State'],values=['S0101_C01_001E (Total_Estimated_Population)','percent_white_population_under_18','Percent_BLACK_OR_AFRICAN_AMERICAN_ALONE_population_under_18','percent_AMERICAN_INDIAN_AND_ALASKA_NATIVE_ALONE_population_under_18','percent_ASIAN_ALONE_population_under_18','percent_NATIVE_HAWAIIAN_AND_OTHER_PACIFIC_ISLANDER_ALONE_population_under_18','percent_SOME_OTHER_RACE_ALONE_population_under_18','percent_TWO_OR_MORE_RACES_population_under_18','percent_WHITE_ALONE_NOT_HISPANIC_OR_LATINO_population_under_18','percent_HISPANIC_OR_LATINO_population_under_18'],columns='Year').reset_index()
df4.head()
```

```python
#storing the data present in the dataframe in a csv while
df4.to_csv('ACS_Race_population_Under18.csv',index=False)
```
