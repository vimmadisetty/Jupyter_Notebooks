```python
import pandas as pd
import csv
import numpy as np
```

```python
#reading the CSV files
#df1 dataframe has the data from 2010 to 2020 for which we have to update the inflation values
#df2 dataframe has inflation values for each year
df1=pd.read_csv('ASM_3_digit_manufacturing_Inflation_rate_update\ASM_3_Manufacturing_EMP,PAYANN,ValueAdd_from_2010_to_2020.csv')
df2=pd.read_csv('ASM_3_digit_manufacturing_Inflation_rate_update\Inflation_value.csv')
#updating the PAYANN value with the respective year inflation rate for example the dollar value in 2010 is not same in 2020 so we need to update the value of dollar with the inflation vlaue
#here we are creating a new column with the name 'Payann_updated' which stores the updated values of payann
# for example in 2010 the PAYANN value is 916514 we have to multiply with respective inflation value 1.185 to obtain the 'Payann_updated'
# similarly this step repeats for all the corresponding maching years
df1['Payann_updated'] = df1.PAYANN.mul(df1.YEAR.map(df2.set_index('YEAR').Value))
# here we are upating the Valueadd with the updated inflation value
df1['Valueadd_updated']=df1.VALADD.mul(df1.YEAR.map(df2.set_index('YEAR').Value))
df1.head()
```

```python
#we are groupingby NAICS title so that we can sum the payann values for each manuufacturing units
a=df1.groupby('NAICS_TTL')['Payann_updated'].sum().sort_values(ascending=False)
a.head(1000)
```
